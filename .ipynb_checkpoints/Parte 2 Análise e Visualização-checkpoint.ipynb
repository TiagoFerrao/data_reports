**Heatmap dos Valores Omissos**

A limpeza e tratamento de dados efetuados na 
[Parte 1](./Parte1%20Limpeza%20e%20Tratamento%20de%20Dados.ipynb)
deste projecto, permitiu eliminar a maior parte de valores omissos, excepto (como se pode ver no heatmap seguinte) um numero significativo de valores nulos na coluna 'c_inc_pos_support?' que corresponde  à pergunta: A incubadora continua a apoiar após a sua saída da incubadora?

- Entre a pergunta do suporte da incubadora e a sua ligação a uma universidade
- Entre os resultados 

```python
#Mapa de correlações 
f, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(survey.corr(), annot=False, square=False, linewidths=.5, ax=ax, vmin=.0)
plt.show()
```

Cálculo da correlação (de Pearson) entre o tempo de residencia e a dimensão da empresa

```python
#cria uma novacoluna onde os valores de dimensao sao numéricos para permitir correlação
replace_keys = {'Micro':0, 'Pequena':1}
survey['c_size_coded']=survey['c_size'].replace(replace_keys)
survey.corr()['c_resident_years']['c_size_coded']
```

<div class="alert alert-block alert-info">
Verifica-se uma maior frequência de empresas com idade entre os 3 e 5 anos de idade com uma dimensão de micro. O número de empresas com mais de 5 anos é reduzida indicando um possível threeshold the mortalidade ou saída da incubadora (corroborado pelo tempo de residência médio de 1,5 anos). Não foi encontrada uma correlação significativa entre o tempo de residência e a dimensão da empresa, ou seja não é possível determinar se o tempo que a empresa passa na incubadora a ajuda a crescer em dimensão. 
</div>

#### 1.3.5.  Fase de incubação  [c_stage] 
A maior parte das empresas encontram-se em fase de incubação.

```python
survey['c_stage' ].value_counts()
```

```python
#estágio de incubação
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.countplot(x='c_stage', data=survey, order=['Graduação','Pré - Incubação','Incubação','Pós - Incubação'])
plt.title('Frequência das empresas por Fase de Incubação')
plt.xlabel('Fases de Incubação')
plt.xticks(rotation=0)
plt.ylabel('Frequência')

plt.subplot(1,2,2)
survey['c_stage'].value_counts().plot(kind='pie',autopct='%1.0f%%', pctdistance=0.7)
plt.title('Percentagem de empresas em cada Fase de Incubação')
plt.ylabel('')
plt.show()
```

Abaixo a relação entre o tempo em que a empresa está na incubadora, e o seu estágio de incubação.

```python
#relaçao entre estagio de incubaçao e tempo na incubadora
plt.figure(figsize=(15,5))
sns.set_theme()
order=["Graduação", "Pré - Incubação", "Incubação","Pós - Incubação"]
sns.stripplot(data=survey, x='c_stage',y='c_resident_years', order=order, hue='c_age', s=10)
plt.title('Distribuição do Tempo de incubação por Fase de Incubação')
plt.xlabel('Fase de Incubação')
plt.xticks(rotation=0)
plt.ylabel('Tempo de Residência na Incubadora')

plt.show()
```
