#### 1.3.5.  Fase de incubao  [c_stage]

```python
survey['c_stage' ].value_counts()
```

```python
#estgio de incubao
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.countplot(x='c_stage', data=survey, order=['Graduao','Pr - Incubao','Incubao','Ps - Incubao'])
plt.title('Frequncia das empresas por Fase de Incubao')
plt.xlabel('Fases de Incubao')
plt.xticks(rotation=0)
plt.ylabel('Frequncia')

plt.subplot(1,2,2)
survey['c_stage'].value_counts().plot(kind='pie',autopct='%1.0f%%', pctdistance=0.7)
plt.title('Percentagem de empresas em cada Fase de Incubao')
plt.ylabel('')
plt.show()
```

Abaixo a relao entre o tempo em que a empresa est na incubadora, e o seu estgio de incubao.

```python
#relaao entre estagio de incubaao e tempo na incubadora
plt.figure(figsize=(15,5))
sns.set_theme()
order=["Graduao", "Pr - Incubao", "Incubao","Ps - Incubao"]
sns.stripplot(data=survey, x='c_stage',y='c_resident_years', order=order, hue='c_age', s=10)
plt.title('Distribuio do Tempo de incubao por Fase de Incubao')
plt.xlabel('Fase de Incubao')
plt.xticks(rotation=0)
plt.ylabel('Tempo de Residncia na Incubadora')

plt.show()
```
